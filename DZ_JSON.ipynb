# 1 Сколько твитов в наборе?
Прочитал файл, сделал массив, посчитал длину 

```python

import json
import re
from collections import Counter
from string import punctuation   
twitter = []
for line in open('hw_3_twitter.json'):       # читаю JSON
    twitter.append(json.loads(line))
print('Amount of tweets:', len(twitter))
```

# 2 Какой процент твитов составляют удаленные записи? (помеченные как deleted)
Ищу по тегу delete, набиваю их в массив и считаю процент

```python
deleted_tweets = []
for tweet in twitter:
    if tweet.get('delete') is not None:  # смотрю наличие тега delete и отбираю по нему в новый массив
        deleted_tweets.append(tweet)
percento = (len(deleted_tweets) / len(twitter)) * 100
print('Deleted tweets:', round(percento, 1), '%')
```

# 3 Какие самые популярные языки твитов?
Ищу тег lang, беру значения и набиваю их в массив, потом делаю Counter и красиво вывожу

```python
langs = []
for tweet in twitter:
    lang = tweet.get('lang') # беру значение по тегу lang 
    if lang is not None:
        langs.append(lang) # заношу в массив всех языков
c_langs = Counter(langs) # делаю Counter по массиву языков
print('TOP-15 languages:')
for i in c_langs.most_common(15):
    print(i[0], ':', i[1], 'tweets')
```

# 4 Есть ли твиты от одного и того же пользователя? Если да, то сколько таких пользователей?
Ищу id каждого пользователя, делаю из id и его followers_count словарь, где ключ - id, а значение - количество твитов с этим id внутри, потом делаю Counter и добавляю id с 2 или более твитами в массив, вывожу длину массива

```python
ids = []
ids_more = []
for tweet in twitter:
    userinfo = tweet.get('user')       # беру user, вынимаю оттуда id
    if userinfo is not None:
        ids.append(userinfo.get('id')) # делаю словарь, где id - ключ, кол-во твитов - значение
c_ids = Counter(ids)
for i in c_ids.most_common():
    if i[1] > 1:               # считаю сколько id со значением >1
        ids_more.append(i)
print('Users with more than one tweet:')
print(len(ids_more))
```

# 5 Топ-20 хэштегов (для них есть специальное поле)
Ищу по ключу entitites, вынимаю оттуда hashtags, потом беру только text хештега, набиваю их в массив, делаю Counter и красиво вывожу

```python
hashtags = []
for tweet in twitter:
    ent = tweet.get('entities')
    if ent is not None:
        h = ent.get('hashtags')   # вынимаю из entities хештеги
        for tag in h:
            hashtag = tag.get('text') # беру только текст хештега и делаю из них массив 
            hashtags.append(hashtag)
c_hashtags = Counter(hashtags)
top = c_hashtags.most_common(20)
print('TOP-20 hashtags:')
for i in top:
    print(i[0], ':', i[1], 'times used')
```

# 6 Предобработать тексты оригинальных твитов (не ретвитов) на английском языке (убрать пунктуацию, привести к нижнему регистру) и составить частотный словарь
Нахожу только НЕ ретвиты на английском, набиваю их в массив, потом чищу от пунктуации и эмодзи, а тексты делаю одной большой строкой, чтобы сплитить по пробелам и сделать из нее массив. Потом делаю Counter для частотного словаря

```python
texts = []
for tweet in twitter:
    if tweet.get('retweeted') is False and tweet.get('lang') == 'en':  # беру НЕ ретвиты на английском языке
        ext = tweet.get('extended_tweet')
        if ext is None:                 # проверяю, есть ли у твита extented
            texts.append(tweet.get('text')) # если нет - беру просто текст
        else:
            texts.append(ext.get('full_text')) # если есть - беру full text из extended
new_text = ''
norm_texts = []
for text in texts:
    text = text.lower()
    for i in punctuation:         # убираю пунктуацию
        text = text.replace(i, '')
    text = text.encode('ascii', 'ignore').decode('ascii') # убираю эмодзи
    norm_texts.append(text)
    new_text += text         # делаю все одной большой строкой
print('Normalized texts:')
for text in norm_texts:
    print(text)
new_text = new_text.split() # спличу большую строку на слова 
words_d = Counter(new_text)
print('Frequency dictionary:')
print(words_d)
```

# 7 Найти количество подписчиков (фолловеров) у авторов твитов и вывести топ-10
Ищу user, оттуда беру name и followers_count, делаю из них словарь (name - ключ, followers_count - значение), делаю Counter и вывожу 10 самых популярных

```python
followers_d = {}
for tweet in twitter:
    userinfo = tweet.get('user')
    if userinfo is not None:
        username = userinfo.get('name')     # делаю словарь по username и followers_count
        followers = userinfo.get('followers_count')
        followers_d[username] = followers
c_followers = Counter(followers_d)
print('TOP-10 followers number')
for i in c_followers.most_common(10):
    print(i[0], ':', i[1], 'followers')
```

# 8 Топ-10 источников твита (из какого приложения написан) (подсказка: для обработки используйте регулярные выражения)
Ищу тег source, регуляркой достаю название приложения, набиваю их в массив и делаю Counter

```python
source_a = []
for tweet in twitter:
    source = tweet.get('source')
    if source is not None:
        src = re.search(r'<a.*>(.*)</a>', source) # по ключу source ищу регуляркой название приложения
        source_a.append(src.group(1))
c_source = Counter(source_a)
print('TOP-10 Twitter apps:')
for i in c_source.most_common(10):
    print(i[0], ':', i[1], 'tweets')
```
